```
import pandas as pd
import re

df = pd.read_csv('/content/nlp_samples_labeled.csv')
df
```

```
def convert_textblob(sentiment):
  polarity = re.search('polarity=..',sentiment)[0]
  polarity = polarity[-2::]
  if float(polarity) > 0:
    return 'Positive'
  elif float(polarity) < 0:
    return 'Negative'
  else:
    return 'Neutral'

def convert_google(sentiment):
  polarity = re.search('.*,',sentiment)[0]
  polarity = polarity[:-1]
  if float(polarity) > 0:
    return 'Positive'
  elif float(polarity) < 0:
    return 'Negative'
  else:
    return 'Neutral'
```

```
convert_google('-0.699999988079071,0.699999988079071')
convert_textblob('Sentiment(polarity=0.0, subjectivity=0.0666666')
```

```
df.rename(columns={'textblob_sentiment':'textblob', 'google_sentiment':'google', 'Researcher_label':'researcher_label', 'text':'tweet'}, inplace=True)

df['textblob_sentiment'] = df.apply(lambda row: convert_textblob(row['textblob']), axis=1)
df['google_sentiment'] = df.apply(lambda row: convert_google(row['google']), axis=1)

df
```

```
df.drop(['google', 'textblob'], inplace=True, axis = 1)
df
```

```
correct_google = len(df[df['google_sentiment']==df['researcher_label']])
correct_textblob = len(df[df['textblob_sentiment']==df['researcher_label']])
correct_nltk = len(df[df['nltk_sentiment']==df['researcher_label']])
```

```
print('Google accuracy:', correct_google/len(df))
print('Textblob accuracy:', correct_textblob/len(df))
print('NLTK accuracy:', correct_nltk/len(df))
```
